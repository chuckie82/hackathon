```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tifffile as tif
import numpy as np
```

```python
import glob
import matplotlib.pyplot as plt
import os
path = input('''This is for X-ray data set. Please copy and paste path to your local data dir:
blank will be set to current directory  ''')
if path in (''):
    path = os.getcwd()
os.chdir(path)
std_list = glob.glob('./*standar*.tif')
dark_list = []
for el in std_list:
    if 'dark' in el:
        dark_list.append(el)
        std_list.remove(el)
print('std_list = {}'.format(std_list))
print('dark_list = {}'.format(dark_list))
numImages = len(std_list)
print("number of images: ",numImages)
```

```python
labels = np.zeros((numImages,))
```

```python
ind = 10 # change this value
labels[ind] = 1 # set label 0:none, 1:excellent, 2:good, 3:patchy, 4:bad
original_img = tif.imread(std_list[ind])
maxVal = np.max(original_img)*0.4
plt.figure(figsize=(10,7))
plt.imshow(original_img[1700::,700:1150], interpolation='none', vmax=maxVal, vmin=0)
plt.title(std_list[ind].split('/')[-1]+": "+str(labels[ind]))
plt.show()
```

```python
print(labels)
myLabels = np.reshape(labels,(len(labels),1,1,1))
np.save('/Users/yoon82/Documents/hackathon/timothy/standard_labels',myLabels)
```
